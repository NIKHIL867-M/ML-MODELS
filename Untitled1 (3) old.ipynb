```python
import os
import pandas as pd

# Load user table first (smaller, main index)
base_dir = '/content/dataset/dataset'
train_dir = os.path.join(base_dir, 'train')

def load_csv(path):
    return pd.read_csv(path)

users = load_csv(os.path.join(train_dir, 'users.csv'))

# Define a function to aggregate big tables by user_id
def aggregate_large_table(path, numeric_cols=None, extra_aggregations=None):
    df = pd.read_csv(path)
    if not numeric_cols:
        numeric_cols = df.select_dtypes(include='number').columns.tolist()
        if 'user_id' in numeric_cols:
            numeric_cols.remove('user_id')
    group = df.groupby('user_id')[numeric_cols].agg(['mean', 'std', 'min', 'max'])
    group.columns = ['_'.join(col).strip() for col in group.columns.values]
    group = group.reset_index()
    # Optionally add more aggregations if needed
    if extra_aggregations:
        for col, agg in extra_aggregations.items():
            group[f'{col}_{agg}'] = df.groupby('user_id')[col].agg(agg).values
    return group

# Example: Aggregate wearables and digital_behavior
wearables_path = os.path.join(train_dir, 'wearables.csv')
digital_behavior_path = os.path.join(train_dir, 'digital_behavior.csv')
wearables_agg = aggregate_large_table(wearables_path)
digital_behavior_agg = aggregate_large_table(digital_behavior_path)

# Load and join other small tables normally (current_products, claims_history, etc)
current_products = load_csv(os.path.join(train_dir, 'current_products.csv'))
claims_history = load_csv(os.path.join(train_dir, 'claims_history.csv'))

# Merge all with users by user_id (keep as left join to preserve all users)
master_train = users.merge(wearables_agg, on='user_id', how='left')
master_train = master_train.merge(digital_behavior_agg, on='user_id', how='left')
master_train = master_train.merge(current_products, on='user_id', how='left')
master_train = master_train.merge(claims_history, on='user_id', how='left')
print("Final master train shape:", master_train.shape)
```

```python
import os
import pandas as pd
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, StackingRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import xgboost as xgb
import joblib

# --- DATA LOADING & MERGING ---
base_dir = '/content/dataset/dataset'
train_dir = os.path.join(base_dir, 'train')

def aggregate_large_table(path):
    df = pd.read_csv(path)
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    if 'user_id' in numeric_cols:
        numeric_cols.remove('user_id')
    if not numeric_cols:
        return pd.DataFrame({'user_id': df['user_id'].unique()})
    group = df.groupby('user_id')[numeric_cols].agg(['mean', 'std', 'min', 'max'])
    group.columns = ['_'.join(col).strip() for col in group.columns.values]
    return group.reset_index()

print("Loading data...")
users = pd.read_csv(os.path.join(train_dir, 'users.csv'))
wearables = aggregate_large_table(os.path.join(train_dir, 'wearables.csv'))
digital = aggregate_large_table(os.path.join(train_dir, 'digital_behavior.csv'))
current_products = pd.read_csv(os.path.join(train_dir, 'current_products.csv'))
claims_history = pd.read_csv(os.path.join(train_dir, 'claims_history.csv'))

master = users.merge(wearables, on='user_id', how='left') \
              .merge(digital, on='user_id', how='left') \
              .merge(current_products, on='user_id', how='left') \
              .merge(claims_history, on='user_id', how='left')

# ===== USE ACTUAL TARGET: claim_amount (REGRESSION) =====
target = 'claim_amount'
X = master.drop(columns=[target, 'user_id', 'claim_id', 'claim_date', 'was_fraudulent'], errors='ignore')
y = master[target].fillna(0)

# Feature types
numeric_features = X.select_dtypes(include=[np.number, 'bool']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

# Preprocessors
numeric_transformer = Pipeline([('imputer', SimpleImputer(strategy='median'))])
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# Train-val split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=42)

# ===== TRAIN REGRESSION MODELS =====
print("Training models...")
lgb_model = lgb.LGBMRegressor(n_estimators=300, max_depth=12, learning_rate=0.05, random_state=42, n_jobs=-1)
rf_model = RandomForestRegressor(n_estimators=200, max_depth=12, random_state=42, n_jobs=-1)
xgb_model = xgb.XGBRegressor(n_estimators=200, max_depth=8, learning_rate=0.05, random_state=42, n_jobs=-1)

# Build pipelines
pipe_lgb = Pipeline([('preprocessor', preprocessor), ('regressor', lgb_model)])
pipe_rf = Pipeline([('preprocessor', preprocessor), ('regressor', rf_model)])
pipe_xgb = Pipeline([('preprocessor', preprocessor), ('regressor', xgb_model)])

pipe_lgb.fit(X_train, y_train)
pipe_rf.fit(X_train, y_train)
pipe_xgb.fit(X_train, y_train)

# ===== STACKING ENSEMBLE (REGRESSION) =====
estimators = [('lgb', pipe_lgb), ('rf', pipe_rf), ('xgb', pipe_xgb)]
stacking_model = StackingRegressor(
    estimators=estimators,
    final_estimator=Ridge(alpha=1.0),
    cv=5,
    n_jobs=-1
)

stacking_model.fit(X_train, y_train)

# Validation RMSE
y_pred = stacking_model.predict(X_val)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print(f" Validation RMSE: {rmse:.4f}")

# Save the stacking model
joblib.dump(stacking_model, 'advanced_stacking_regressor.joblib')
print(" Advanced stacking regressor saved!")
```

```python
import os
import pandas as pd
import numpy as np
import lightgbm as lgb
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import joblib
import matplotlib.pyplot as plt
import shap
import warnings
warnings.filterwarnings('ignore')

# ========== PHASE 1: LOAD & PREPARE DATA ==========
base_dir = '/content/dataset/dataset'
train_dir = os.path.join(base_dir, 'train')

def aggregate_large_table(path):
    df = pd.read_csv(path)
    numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
    if 'user_id' in numeric_cols:
        numeric_cols.remove('user_id')
    if not numeric_cols:
        return pd.DataFrame({'user_id': df['user_id'].unique()})
    group = df.groupby('user_id')[numeric_cols].agg(['mean', 'std', 'min', 'max'])
    group.columns = ['_'.join(col).strip() for col in group.columns.values]
    return group.reset_index()

print("=== LOADING DATA ===")
users = pd.read_csv(os.path.join(train_dir, 'users.csv'))
wearables = aggregate_large_table(os.path.join(train_dir, 'wearables.csv'))
digital = aggregate_large_table(os.path.join(train_dir, 'digital_behavior.csv'))
current_products = pd.read_csv(os.path.join(train_dir, 'current_products.csv'))
claims_history = pd.read_csv(os.path.join(train_dir, 'claims_history.csv'))

master = users.merge(wearables, on='user_id', how='left') \
              .merge(digital, on='user_id', how='left') \
              .merge(current_products, on='user_id', how='left') \
              .merge(claims_history, on='user_id', how='left')

# ========== PHASE 2: ADVANCED FEATURE ENGINEERING ==========
print("=== FEATURE ENGINEERING ===")
target = 'claim_amount'
X = master.drop(columns=[target, 'user_id', 'claim_id', 'claim_date', 'was_fraudulent'], errors='ignore')
y = master[target].fillna(0)

# Add engineered features
X['claim_frequency'] = master.groupby('user_id')['claim_id'].transform('count')
X['income_bucket'] = pd.qcut(X['annual_income'], 5, labels=False, duplicates='drop') if 'annual_income' in X.columns else 0
X['lifestyle_score'] = X.select_dtypes(include=['number']).mean(axis=1)

numeric_features = X.select_dtypes(include=[np.number, 'bool']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

numeric_transformer = Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# ========== PHASE 3: HYBRID MODELS (SVM, NB + Advanced Models) ==========
print("=== TRAINING HYBRID MODELS ===")
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=42)

# Advanced models
lgb_model = lgb.LGBMRegressor(n_estimators=300, max_depth=12, learning_rate=0.05, random_state=42, n_jobs=-1)
rf_model = __import__('sklearn.ensemble', fromlist=['RandomForestRegressor']).RandomForestRegressor(n_estimators=200, max_depth=12, random_state=42, n_jobs=-1)
xgb_model = xgb.XGBRegressor(n_estimators=200, max_depth=8, learning_rate=0.05, random_state=42, n_jobs=-1)

# Special models: SVM + Naive Bayes
svm_model = SVR(kernel='rbf', C=100, gamma=0.1)
nb_model = GaussianNB()

# Deep model: Neural Network
deep_model = MLPRegressor(hidden_layer_sizes=(128, 64), max_iter=500, random_state=42, early_stopping=True)

# Build pipelines for each model
pipe_lgb = Pipeline([('preprocessor', preprocessor), ('model', lgb_model)])
pipe_xgb = Pipeline([('preprocessor', preprocessor), ('model', xgb_model)])
pipe_svm = Pipeline([('preprocessor', preprocessor), ('model', svm_model)])
pipe_deep = Pipeline([('preprocessor', preprocessor), ('model', deep_model)])

# Train all base models
print("Training LightGBM, XGBoost, SVM, Deep models...")
pipe_lgb.fit(X_train, y_train)
pipe_xgb.fit(X_train, y_train)
pipe_svm.fit(X_train, y_train)
pipe_deep.fit(X_train, y_train)

# ========== PHASE 4: STACKING ENSEMBLE (META-MODEL) ==========
print("=== STACKING ENSEMBLE ===")
estimators = [
    ('lgb', pipe_lgb),
    ('xgb', pipe_xgb),
    ('svm', pipe_svm),
    ('deep', pipe_deep)
]

stacking_model = StackingRegressor(
    estimators=estimators,
    final_estimator=Ridge(alpha=1.0),
    cv=5,
    n_jobs=-1
)

stacking_model.fit(X_train, y_train)

# Validation performance
y_pred_stack = stacking_model.predict(X_val)
rmse_stack = np.sqrt(mean_squared_error(y_val, y_pred_stack))
r2_stack = r2_score(y_val, y_pred_stack)
print(f" Stacking RMSE: {rmse_stack:.4f}, R2: {r2_stack:.4f}")

# ========== PHASE 5: FEATURE IMPORTANCE REPORT ==========
print("=== FEATURE IMPORTANCE ANALYSIS ===")
lgb_importance = pd.DataFrame({
    'feature': numeric_features + categorical_features,
    'importance': pipe_lgb.named_steps['model'].feature_importances_[:len(numeric_features + categorical_features)]
}).sort_values('importance', ascending=False).head(15)

print("\nTop 15 Important Features:")
print(lgb_importance)

# Save feature importance report
lgb_importance.to_csv('feature_importance_report.csv', index=False)
print(" Feature importance saved to feature_importance_report.csv")

# ========== PHASE 6: INSURANCE RISK SEGMENTATION ==========
print("=== INSURANCE RISK SEGMENTATION ===")
y_pred_all = stacking_model.predict(X)
risk_quantiles = pd.qcut(y_pred_all, 4, labels=['Low', 'Medium', 'High', 'Very High'])

risk_segment_df = pd.DataFrame({
    'user_id': master['user_id'].values[:len(risk_quantiles)],
    'predicted_claim': y_pred_all,
    'risk_segment': risk_quantiles
})

print("\nRisk Segmentation Summary:")
print(risk_segment_df['risk_segment'].value_counts())
risk_segment_df.to_csv('insurance_risk_segmentation.csv', index=False)
print(" Risk segmentation saved to insurance_risk_segmentation.csv")

# ========== PHASE 7: SHAP EXPLAINABILITY ==========
print("=== SHAP EXPLAINABILITY ===")
try:
    # Use SHAP for model interpretation
    explainer = shap.TreeExplainer(pipe_lgb.named_steps['model'])
    X_sample = preprocessor.transform(X_val.head(100))
    shap_values = explainer.shap_values(X_sample)

    # Save SHAP plot
    plt.figure(figsize=(12, 6))
    shap.summary_plot(shap_values, X_sample, show=False)
    plt.savefig('shap_summary_plot.png', dpi=100, bbox_inches='tight')
    print(" SHAP summary plot saved to shap_summary_plot.png")

except Exception as e:
    print(f" SHAP visualization limited: {e}")

# ========== SAVE ALL MODELS ==========
print("=== SAVING ALL MODELS ===")
joblib.dump(stacking_model, 'upgraded_stacking_model.joblib')
joblib.dump(preprocessor, 'upgraded_preprocessor.joblib')
print(" All models saved!")

print("\n" + "="*50)
print(" UPGRADE PHASE COMPLETE!")
print("="*50)
print("\nGenerated Outputs:")
print("  1. feature_importance_report.csv")
print("  2. insurance_risk_segmentation.csv")
print("  3. shap_summary_plot.png")
print("  4. upgraded_stacking_model.joblib")
print("  5. upgraded_preprocessor.joblib")
```

```python
# =========================
# FrequencySeverity Ensemble (CPU-only, Colab-safe, no-issues)
# =========================

import os, gc, ast, warnings, inspect
import numpy as np
import pandas as pd
import joblib
import lightgbm as lgb
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

warnings.filterwarnings("ignore")

# --- CONFIG ---
BASE_DIR = '/content/dataset/dataset'
TRAIN_DIR = os.path.join(BASE_DIR, 'train')
TEST_DIR  = os.path.join(BASE_DIR, 'test')

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# ---------------------------
# Utils / IO
# ---------------------------
def load_csv(path):
    return pd.read_csv(path, encoding='latin-1')

def _safe_get(df, col, default=0):
    return df[col] if col in df.columns else default

# ---------------------------
# Feature engineering
# ---------------------------
def aggregate_large_table(path, name):
    df = load_csv(path)
    if 'user_id' not in df.columns:
        raise ValueError(f"'user_id' not found in {path}")

    numeric_cols = df.select_dtypes(include='number').columns.tolist()
    if 'user_id' in numeric_cols:
        numeric_cols.remove('user_id')

    if not numeric_cols:
        return pd.DataFrame({'user_id': df['user_id'].unique()})

    group = df.groupby('user_id')[numeric_cols].agg(['mean','std','min','max'])
    group.columns = [f'{name}_{"_".join(col).strip()}' for col in group.columns.values]
    group = group.reset_index()

    # Telematics extras
    if name == 'telematics':
        hb = 'hard_braking_events' in df.columns
        sp = 'speeding_events' in df.columns
        if hb or sp:
            df['hard_braking_events'] = _safe_get(df, 'hard_braking_events', 0)
            df['speeding_events']     = _safe_get(df, 'speeding_events', 0)
            df['total_events']        = df['hard_braking_events'] + df['speeding_events']
            events_mean = df.groupby('user_id')['total_events'].mean().reset_index()
            events_mean.columns = ['user_id','telematics_events_per_trip_mean']
            group = group.merge(events_mean, on='user_id', how='left')

    return group

def extract_json_features(df, col_name, prefix):
    if col_name not in df.columns:
        return df

    def safe_literal_eval(x):
        if pd.isna(x): return {}
        if isinstance(x, dict): return x
        if isinstance(x, str):
            try: return ast.literal_eval(x)
            except Exception: return {}
        return {}

    s = df[col_name].apply(safe_literal_eval)

    def sum_values(d):
        try:
            return float(sum(v for v in d.values() if pd.notna(v)))
        except Exception:
            return 0.0

    df[f'{prefix}_vector_sum'] = s.apply(sum_values)

    keys = ['US Stocks','International Stocks','Bonds'] if prefix == 'alloc' else []
    for key in keys:
        df[f'{prefix}_{key.lower().replace(" ","_")}'] = s.apply(lambda d: d.get(key, 0))

    return df.drop(columns=[col_name], errors='ignore')

def create_master_df(folder_path, claims_history=None):
    print(f"Loading data from {folder_path}...")
    users       = load_csv(os.path.join(folder_path, 'users.csv'))
    wearables   = aggregate_large_table(os.path.join(folder_path, 'wearables.csv'),           'wearable')
    digital     = aggregate_large_table(os.path.join(folder_path, 'digital_behavior.csv'),    'digital')
    telematics  = aggregate_large_table(os.path.join(folder_path, 'telematics.csv'),          'telematics')

    current_products = load_csv(os.path.join(folder_path, 'current_products.csv'))
    current_products = extract_json_features(current_products, 'portfolio_asset_allocation', 'alloc')

    master = users.merge(wearables,  on='user_id', how='left') \
                  .merge(digital,    on='user_id', how='left') \
                  .merge(telematics, on='user_id', how='left') \
                  .merge(current_products, on='user_id', how='left')

    if claims_history is not None:
        master = master.merge(claims_history, on='user_id', how='left')

    # Guard missing cols for engineered features
    for c in ['annual_income','student_loan_debt','credit_card_debt',
              'risk_tolerance_score','financial_literacy_score']:
        if c not in master.columns: master[c] = np.nan

    master['financial_health'] = master['annual_income'].fillna(0) \
                              - master['student_loan_debt'].fillna(0) \
                              - master['credit_card_debt'].fillna(0)
    master['risk_propensity']  = master['risk_tolerance_score'].fillna(0) \
                               * master['financial_literacy_score'].fillna(0)
    return master

def prepare_targets_and_features(master_train):
    claims_history_sub = master_train.dropna(subset=['claim_id']).copy()
    if 'policy_type' in claims_history_sub.columns:
        claims_history_sub['policy_type'] = claims_history_sub['policy_type'].str.lower()

    claim_targets = claims_history_sub.groupby(['user_id','policy_type']).agg(
        freq=('claim_id','count'),
        sev_base=('claim_amount','mean')
    ).reset_index()

    pivot = claim_targets.pivot_table(index='user_id', columns='policy_type',
                                      values=['freq','sev_base']).reset_index()

    cols = list(pivot.columns)
    if isinstance(cols[0], tuple) and cols[0][0] == 'user_id':
        pivot.rename(columns={cols[0]:'user_id'}, inplace=True)
    elif cols[0] == 'level_0':
        pivot.rename(columns={'level_0':'user_id'}, inplace=True)

    new_cols = ['user_id']
    for col in pivot.columns[1:]:
        new_cols.append(f'{col[0]}_{str(col[1]).lower()}') if isinstance(col, tuple) else new_cols.append(col)
    pivot.columns = new_cols

    target_df = master_train.merge(pivot, on='user_id', how='left', suffixes=('_orig',''))

    for p in ['auto','renters','pet']:
        if f'freq_{p}' not in target_df.columns: target_df[f'freq_{p}'] = 0
        target_df[f'freq_{p}'] = target_df[f'freq_{p}'].fillna(0).astype(float).apply(lambda x: 1.0 if x>0 else 0.0)

        sev_col = f'sev_base_{p}'
        if sev_col not in target_df.columns: target_df[sev_col] = np.nan
        target_df[f'sev_{p}'] = target_df[sev_col].astype(float)

    drop_cols = [
        'user_id','claim_id','claim_date','was_fraudulent','policy_type',
        'sev_base_auto','sev_base_renters','sev_base_pet',
        'freq_auto','freq_renters','freq_pet'
    ]
    X = target_df.drop(columns=[c for c in drop_cols if c in target_df.columns], errors='ignore')
    return X, target_df

# ---------------------------
# Preprocessing (robust to sklearn version)
# ---------------------------
def _make_ohe():
    # sklearn >=1.4 uses sparse_output; older uses sparse.
    sig = inspect.signature(OneHotEncoder.__init__)
    if 'sparse_output' in sig.parameters:
        return OneHotEncoder(handle_unknown='ignore', sparse_output=True)
    else:
        return OneHotEncoder(handle_unknown='ignore', sparse=True)

def get_preprocessor(X):
    X_ = X.copy()
    for c in X_.select_dtypes(include='bool').columns:
        X_[c] = X_[c].astype(int)

    num_cols = X_.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = X_.select_dtypes(include=['object']).columns.tolist()

    num = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])
    cat = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', _make_ohe())
    ])

    return ColumnTransformer(
        transformers=[
            ('num', num, num_cols),
            ('cat', cat, cat_cols)
        ],
        remainder='drop',
        sparse_threshold=0.3
    )

# ---------------------------
# Models (CPU-only)
# ---------------------------
def compute_class_weights(y):
    y = np.asarray(y).astype(float)
    pos = y.sum()
    neg = len(y) - pos
    scale_pos_weight = (neg / max(pos, 1.0))
    cw = None if (pos == 0 or neg == 0) else {0: 1.0, 1: (neg / pos)}
    return scale_pos_weight, cw

def make_lgbm_classifier(class_weight):
    return lgb.LGBMClassifier(
        n_estimators=2000,
        learning_rate=0.02,
        max_depth=12,
        num_leaves=96,
        subsample=0.85,
        colsample_bytree=0.8,
        reg_alpha=0.2,
        reg_lambda=0.8,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        objective='binary',
        metric='auc',
        class_weight=class_weight
    )

def make_xgb_classifier(scale_pos_weight):
    return xgb.XGBClassifier(
        n_estimators=2000,
        learning_rate=0.02,
        max_depth=8,
        subsample=0.85,
        colsample_bytree=0.8,
        reg_lambda=1.2,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        objective='binary:logistic',
        eval_metric='auc',
        tree_method='hist',
        predictor='auto',
        scale_pos_weight=scale_pos_weight
    )

def make_lgbm_regressor():
    return lgb.LGBMRegressor(
        n_estimators=1600,
        learning_rate=0.02,
        max_depth=16,
        num_leaves=128,
        subsample=0.85,
        colsample_bytree=0.8,
        reg_alpha=0.7,
        reg_lambda=1.5,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        objective='regression',
        metric='rmse'
    )

def make_xgb_regressor():
    return xgb.XGBRegressor(
        n_estimators=1600,
        learning_rate=0.02,
        max_depth=10,
        subsample=0.85,
        colsample_bytree=0.8,
        reg_lambda=2.0,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        objective='reg:squarederror',
        eval_metric='rmse',
        tree_method='hist',
        predictor='auto'
    )

# LightGBM older versions don't accept verbose kwarg; silence via callback
def fit_with_holdout(model, Xtr, ytr, Xva, yva):
    if isinstance(model, (lgb.LGBMClassifier, lgb.LGBMRegressor)):
        model.fit(
            Xtr, ytr,
            eval_set=[(Xva, yva)],
            callbacks=[lgb.log_evaluation(0)]  # no eval logging
        )
    else:
        model.fit(
            Xtr, ytr,
            eval_set=[(Xva, yva)],
            verbose=False
        )
    return model

def train_pair_ensemble(X, y, preprocessor, is_classifier: bool):
    X_proc = preprocessor.transform(X)

    # stratify only when both classes are present (avoids ValueError)
    strat = y if (is_classifier and len(np.unique(y)) > 1) else None
    X_tr, X_va, y_tr, y_va = train_test_split(
        X_proc, y, test_size=0.1, random_state=RANDOM_STATE, stratify=strat
    )

    models = {}
    if is_classifier:
        spw, cw = compute_class_weights(y)
        lgbm = make_lgbm_classifier(cw)
        xgbc = make_xgb_classifier(spw)
        models['LGBM'] = fit_with_holdout(lgbm, X_tr, y_tr, X_va, y_va)
        models['XGB']  = fit_with_holdout(xgbc, X_tr, y_tr, X_va, y_va)
    else:
        lgbr = make_lgbm_regressor()
        xgbr = make_xgb_regressor()
        models['LGBM'] = fit_with_holdout(lgbr, X_tr, y_tr, X_va, y_va)
        models['XGB']  = fit_with_holdout(xgbr, X_tr, y_tr, X_va, y_va)
    return models

def predict_ensemble(models, X, preprocessor, is_classifier):
    Xp = preprocessor.transform(X)
    out = []
    for m in models.values():
        if is_classifier and hasattr(m, "predict_proba"):
            out.append(m.predict_proba(Xp)[:,1])
        else:
            out.append(m.predict(Xp))
    return np.mean(out, axis=0)

# ---------------------------
# MAIN
# ---------------------------
if __name__ == "__main__":
    print("="*60)
    print("=== PHASE 1: DATA PREP (FREQUENCY-SEVERITY) ===")
    print("="*60)

    claims_history = load_csv(os.path.join(TRAIN_DIR, 'claims_history.csv'))
    master_train   = create_master_df(TRAIN_DIR, claims_history)
    X_train_full, target_df = prepare_targets_and_features(master_train)

    preprocessor = get_preprocessor(X_train_full)
    preprocessor.fit(X_train_full)

    del master_train
    gc.collect()

    print("\n" + "="*60)
    print("=== PHASE 2: TRAIN ENSEMBLES (CPU-only) ===")
    print("="*60)

    trained = {}
    for disp, p in [('Auto','auto'), ('Renters','renters'), ('Pet','pet')]:
        print(f"\n--- Training {disp} ---")
        y_freq = target_df[f'freq_{p}'].astype(float).values
        print(f"  -> Classifier target: {int(y_freq.sum())}/{len(y_freq)} users claimed.")

        # P* (classifier)
        freq_models = train_pair_ensemble(X_train_full, y_freq, preprocessor, is_classifier=True)
        trained[f'p_{p}'] = freq_models

        # SEV* (regressor)  log1p target for stability; inverse with expm1 at predict time
        sev_mask = (y_freq == 1.0)
        X_sev = X_train_full.loc[sev_mask]
        y_sev = target_df.loc[sev_mask, f'sev_{p}'].astype(float).values
        print(f"  -> Regressor target: Subset Size = {len(y_sev)} users.")
        if len(y_sev) > 0:
            y_sev_log = np.log1p(np.clip(y_sev, a_min=0, a_max=None))
            sev_models = train_pair_ensemble(X_sev, y_sev_log, preprocessor, is_classifier=False)
        else:
            class DummyReg:
                def predict(self, X): return np.zeros(getattr(X, "shape", [len(X)])[0], dtype=float)  # log(1)=0
            sev_models = {'LGBM': DummyReg(), 'XGB': DummyReg()}
        trained[f'sev_{p}'] = sev_models

    joblib.dump(trained, 'ultimate_ensemble_models_cpu.joblib')
    joblib.dump(preprocessor, 'final_preprocessor_cpu.joblib')
    print("\n Saved models + preprocessor (CPU-only).")

    print("\n" + "="*60)
    print("=== PHASE 3: PREDICTION & SUBMISSION ===")
    print("="*60)

    master_test = create_master_df(TEST_DIR)
    if 'user_id' not in master_test.columns:
        raise ValueError("user_id missing in test data")

    X_test = master_test.drop(columns=['user_id'], errors='ignore')
    expected_features = X_train_full.columns.tolist()
    X_test_aligned = X_test.reindex(columns=expected_features, fill_value=np.nan)

    submission = pd.DataFrame({'user_id': master_test['user_id']})

    for p in ['auto','renters','pet']:
        # P*
        p_pred = predict_ensemble(trained[f'p_{p}'], X_test_aligned, preprocessor, is_classifier=True)
        submission[f'p_{p}'] = np.clip(p_pred, 0.001, 0.999)

        # SEV* (inverse of log1p)
        sev_log_pred = predict_ensemble(trained[f'sev_{p}'], X_test_aligned, preprocessor, is_classifier=False)
        sev_pred = np.expm1(sev_log_pred)
        submission[f'sev_{p}'] = np.maximum(1.0, sev_pred)

    cols = ['user_id','p_auto','p_renters','p_pet','sev_auto','sev_renters','sev_pet']
    submission[cols].to_csv('submission_predictions_ultimate.csv', index=False)

    print("\n Created 'submission_predictions_ultimate.csv'. Preview:")
    print(submission[cols].head())
    print("\n Done! (CPU-only, stable across sklearn versions, no GPU/OpenCL issues).")
```

```python
# =========================
# LEAK-PROOF FREQUENCYSEVERITY BAGGED OOF ENSEMBLE (CPU-only, Colab-safe)
# - X is built ONLY from pure feature table (NO targets merged into X)
# - 5x3 bagged OOF CV + version-proof early stopping + isotonic calibration
# - Severity: Tweedie + Pseudo-Huber + Log-MSE stacked with Ridge (OOF)
# - Preflight leakage scan + fold safety stop
# - Submission: user_id, p_auto, p_renters, p_pet, sev_auto, sev_renters, sev_pet
# =========================

import os, gc, ast, json, warnings, inspect
import numpy as np
import pandas as pd
warnings.filterwarnings("ignore")

from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.metrics import roc_auc_score, roc_curve, mean_squared_error
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.isotonic import IsotonicRegression
from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.base import BaseEstimator, TransformerMixin

import lightgbm as lgb
import xgboost as xgb
import joblib

# ---------------------------
# CONFIG
# ---------------------------
BASE_DIR   = '/content/dataset/dataset'
TRAIN_DIR  = os.path.join(BASE_DIR, 'train')
TEST_DIR   = os.path.join(BASE_DIR, 'test')

N_FOLDS      = 5
BAG_SEEDS    = [42, 1337, 2025]
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

LEAKY_NAME_FRAGMENTS = [
    'claim', 'policy_type', 'fraud', 'settle', 'payout', 'accident', 'incident', 'loss',
    # extra belt-and-suspenders:
    'freq_', 'sev_', 'sev_log_', 'target', 'label'
]

# ---------------------------
# IO
# ---------------------------
def load_csv(path):
    return pd.read_csv(path, encoding='latin-1')

def _safe_get(df, col, default=0):
    return df[col] if col in df.columns else default

# ---------------------------
# Simple transformers
# ---------------------------
class Winsorizer(BaseEstimator, TransformerMixin):
    def __init__(self, quantiles=(0.01, 0.99)):
        self.quantiles = quantiles
        self.bounds_ = {}
    def fit(self, X, y=None):
        X_df = pd.DataFrame(X).copy()
        for col in X_df.columns:
            s = X_df[col].astype(float)
            lo = np.nanquantile(s, self.quantiles[0])
            hi = np.nanquantile(s, self.quantiles[1])
            self.bounds_[col] = (lo, hi)
        return self
    def transform(self, X):
        X_df = pd.DataFrame(X).copy()
        for col, (lo, hi) in self.bounds_.items():
            X_df[col] = np.clip(X_df[col], lo, hi)
        return X_df.values

class RareBucket(BaseEstimator, TransformerMixin):
    def __init__(self, min_count=20):
        self.min_count = min_count
        self.maps_ = {}
    def fit(self, X, y=None):
        X_df = pd.DataFrame(X).astype("object")
        for col in X_df.columns:
            vc = X_df[col].value_counts(dropna=False)
            keep = set(vc[vc >= self.min_count].index)
            self.maps_[col] = keep
        return self
    def transform(self, X):
        X_df = pd.DataFrame(X).astype("object")
        for col, keep in self.maps_.items():
            X_df[col] = X_df[col].apply(lambda v: v if v in keep else 'Other')
        return X_df.values

# ---------------------------
# Feature engineering (leak-proof)
# ---------------------------
def extract_json_features(df, col_name, prefix):
    if col_name not in df.columns:
        return df
    def safe_literal_eval(x):
        if pd.isna(x): return {}
        if isinstance(x, dict): return x
        if isinstance(x, str):
            try: return ast.literal_eval(x)
            except Exception: return {}
        return {}
    s = df[col_name].apply(safe_literal_eval)
    def sum_values(d):
        try: return float(sum(v for v in d.values() if pd.notna(v)))
        except Exception: return 0.0
    df[f'{prefix}_vector_sum'] = s.apply(sum_values)
    keys = ['US Stocks','International Stocks','Bonds'] if prefix == 'alloc' else []
    for key in keys:
        df[f'{prefix}_{key.lower().replace(" ","_")}'] = s.apply(lambda d: d.get(key, 0))
    return df.drop(columns=[col_name], errors='ignore')

def aggregate_large_table(path, name):
    df = load_csv(path)
    if 'user_id' not in df.columns:
        raise ValueError(f"'user_id' not found in {path}")
    numeric_cols = df.select_dtypes(include='number').columns.tolist()
    if 'user_id' in numeric_cols: numeric_cols.remove('user_id')
    if not numeric_cols:
        return pd.DataFrame({'user_id': df['user_id'].unique()})
    group = df.groupby('user_id')[numeric_cols].agg(['mean','std','min','max'])
    group.columns = [f'{name}_{"_".join(col).strip()}' for col in group.columns.values]
    group = group.reset_index()
    if name == 'telematics':
        df['hard_braking_events'] = _safe_get(df, 'hard_braking_events', 0)
        df['speeding_events']     = _safe_get(df, 'speeding_events', 0)
        df['total_events']        = df['hard_braking_events'] + df['speeding_events']
        events_mean = df.groupby('user_id')['total_events'].mean().reset_index()
        events_mean.columns = ['user_id','telematics_events_per_trip_mean']
        group = group.merge(events_mean, on='user_id', how='left')
    return group

def create_feature_master(folder_path):
    """Pure features at user level; NO claims merge here."""
    print(f"Loading data from {folder_path}...")
    users = load_csv(os.path.join(folder_path, 'users.csv')).drop_duplicates('user_id', keep='first')
    wearables  = aggregate_large_table(os.path.join(folder_path, 'wearables.csv'),        'wearable')
    digital    = aggregate_large_table(os.path.join(folder_path, 'digital_behavior.csv'), 'digital')
    telematics = aggregate_large_table(os.path.join(folder_path, 'telematics.csv'),       'telematics')
    cur_prod   = load_csv(os.path.join(folder_path, 'current_products.csv'))
    cur_prod   = extract_json_features(cur_prod, 'portfolio_asset_allocation', 'alloc')

    feat = users.merge(wearables,  on='user_id', how='left') \
                .merge(digital,    on='user_id', how='left') \
                .merge(telematics, on='user_id', how='left') \
                .merge(cur_prod,   on='user_id', how='left')

    # engineered
    for c in ['annual_income','student_loan_debt','credit_card_debt',
              'risk_tolerance_score','financial_literacy_score']:
        if c not in feat.columns: feat[c] = np.nan
    feat['financial_health'] = feat['annual_income'].fillna(0) \
                             - feat['student_loan_debt'].fillna(0) \
                             - feat['credit_card_debt'].fillna(0)
    feat['risk_propensity']  = feat['risk_tolerance_score'].fillna(0) \
                             * feat['financial_literacy_score'].fillna(0)

    # hard drop any accidental target-like names
    drop_like = [c for c in feat.columns if any(tok in c.lower() for tok in LEAKY_NAME_FRAGMENTS)]
    if drop_like:
        print(f" Removing potential leak columns from features: {drop_like[:12]}{' ...' if len(drop_like)>12 else ''}")
        feat = feat.drop(columns=drop_like, errors='ignore')
    return feat

def build_targets_from_claims(feature_master, claims_history):
    ch = claims_history.copy()
    if 'policy_type' in ch.columns:
        ch['policy_type'] = ch['policy_type'].str.lower()
    agg = ch.groupby(['user_id','policy_type']).agg(
        freq=('claim_id','count'),
        sev_base=('claim_amount','mean')
    ).reset_index()
    pv = agg.pivot_table(index='user_id', columns='policy_type',
                         values=['freq','sev_base']).reset_index()
    cols = list(pv.columns)
    if isinstance(cols[0], tuple) and cols[0][0] == 'user_id':
        pv.rename(columns={cols[0]:'user_id'}, inplace=True)
    new_cols = ['user_id']
    for col in pv.columns[1:]:
        new_cols.append(f'{col[0]}_{str(col[1]).lower()}') if isinstance(col, tuple) else new_cols.append(col)
    pv.columns = new_cols

    tgt = feature_master[['user_id']].merge(pv, on='user_id', how='left')
    for p in ['auto','renters','pet']:
        if f'freq_{p}' not in tgt.columns: tgt[f'freq_{p}'] = 0
        tgt[f'freq_{p}'] = tgt[f'freq_{p}'].fillna(0).astype(float).gt(0).astype(float)
        sev_col = f'sev_base_{p}'
        if sev_col not in tgt.columns: tgt[sev_col] = np.nan
        tgt[f'sev_log_{p}'] = np.log1p(np.clip(tgt[sev_col].astype(float), 0, None))
        tgt.drop(columns=[sev_col], inplace=True, errors='ignore')
    return tgt

def build_Xy(feature_master, targets):
    """
    X is built ONLY from the pure feature table (no target columns).
    """
    feat_cols = [c for c in feature_master.columns
                 if c != 'user_id' and not any(tok in c.lower() for tok in ['freq_', 'sev_', 'sev_log_'])]
    X = feature_master[feat_cols].copy()

    df_t = targets.copy()
    y_pack = {
        'auto':    df_t.get('freq_auto',    pd.Series(0, index=df_t.index)).astype(float).values,
        'renters': df_t.get('freq_renters', pd.Series(0, index=df_t.index)).astype(float).values,
        'pet':     df_t.get('freq_pet',     pd.Series(0, index=df_t.index)).astype(float).values,
        'sev_log_auto':    df_t.get('sev_log_auto',    pd.Series(np.nan, index=df_t.index)).astype(float).values,
        'sev_log_renters': df_t.get('sev_log_renters', pd.Series(np.nan, index=df_t.index)).astype(float).values,
        'sev_log_pet':     df_t.get('sev_log_pet',     pd.Series(np.nan, index=df_t.index)).astype(float).values,
    }
    return X, y_pack


# ---------------------------
# Preprocessor (sklearn-version-proof)
# ---------------------------
def _make_ohe(min_freq=20):
    sig = inspect.signature(OneHotEncoder.__init__)
    kwargs = {}
    if 'min_frequency' in sig.parameters:
        kwargs['handle_unknown'] = 'infrequent_if_exist'
        kwargs['min_frequency']  = min_freq
    else:
        kwargs['handle_unknown'] = 'ignore'
    if 'sparse_output' in sig.parameters:
        kwargs['sparse_output'] = True
    else:
        kwargs['sparse'] = True
    return OneHotEncoder(**kwargs)

def get_preprocessor(X):
    X_ = X.copy()
    for c in X_.select_dtypes(include='bool').columns:
        X_[c] = X_[c].astype(int)
    num_cols = X_.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = X_.select_dtypes(include=['object']).columns.tolist()

    num = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('winsor',  Winsorizer((0.01, 0.99))),
        ('scaler',  StandardScaler())
    ])

    sig = inspect.signature(OneHotEncoder.__init__)
    cat_steps = [('imputer', SimpleImputer(strategy='most_frequent'))]
    if 'min_frequency' not in sig.parameters:
        cat_steps.append(('rare', RareBucket(min_count=20)))
    cat_steps.append(('onehot', _make_ohe(20)))
    cat = Pipeline(steps=cat_steps)

    return ColumnTransformer(
        transformers=[('num', num, num_cols), ('cat', cat, cat_cols)],
        remainder='drop',
        sparse_threshold=0.3
    )

# ---------------------------
# Models + ES (version-proof)
# ---------------------------
def compute_class_weights(y):
    y = np.asarray(y).astype(float)
    pos = y.sum()
    neg = len(y) - pos
    scale_pos_weight = (neg / max(pos, 1.0))
    cw = None if (pos == 0 or neg == 0) else {0: 1.0, 1: (neg / pos)}
    return scale_pos_weight, cw

def make_lgbm_classifier(cw):
    return lgb.LGBMClassifier(
        n_estimators=6000, learning_rate=0.008,
        max_depth=7, num_leaves=48,
        min_child_samples=80,
        subsample=0.8, colsample_bytree=0.7,
        reg_alpha=0.3, reg_lambda=1.5,
        objective='binary', metric='auc', class_weight=cw,
        n_jobs=-1, random_state=RANDOM_STATE, verbose=-1
    )

def make_xgb_classifier(spw):
    return xgb.XGBClassifier(
        n_estimators=6000, learning_rate=0.008,
        max_depth=6, min_child_weight=6,
        subsample=0.8, colsample_bytree=0.7,
        reg_lambda=2.0, reg_alpha=0.0,
        gamma=0.15,
        objective='binary:logistic', eval_metric='auc',
        tree_method='hist', predictor='auto',
        n_jobs=-1, random_state=RANDOM_STATE,
        scale_pos_weight=spw
    )

def make_lgbm_regressor_tweedie():
    return lgb.LGBMRegressor(
        n_estimators=6000, learning_rate=0.008,
        max_depth=8, num_leaves=96,
        min_child_samples=50, subsample=0.8, colsample_bytree=0.7,
        reg_alpha=0.3, reg_lambda=1.5,
        objective='tweedie', tweedie_variance_power=1.3,
        metric='rmse',
        n_jobs=-1, random_state=RANDOM_STATE, verbose=-1
    )

def make_lgbm_regressor_log():
    return lgb.LGBMRegressor(
        n_estimators=6000, learning_rate=0.008,
        max_depth=8, num_leaves=96,
        min_child_samples=50, subsample=0.8, colsample_bytree=0.7,
        reg_alpha=0.3, reg_lambda=1.5,
        objective='regression', metric='rmse',
        n_jobs=-1, random_state=RANDOM_STATE, verbose=-1
    )

def make_xgb_regressor_phuber():
    return xgb.XGBRegressor(
        n_estimators=6000, learning_rate=0.008,
        max_depth=7, min_child_weight=6,
        subsample=0.8, colsample_bytree=0.7,
        reg_lambda=2.5, reg_alpha=0.0, gamma=0.1,
        objective='reg:pseudohubererror', eval_metric='rmse',
        tree_method='hist', predictor='auto',
        n_jobs=-1, random_state=RANDOM_STATE
    )

def fit_lgbm_es(model, X_tr, y_tr, X_va, y_va):
    model.fit(
        X_tr, y_tr,
        eval_set=[(X_va, y_va)],
        callbacks=[lgb.log_evaluation(0), lgb.early_stopping(stopping_rounds=300, verbose=False)]
    )
    return model

def fit_xgb_es(model, X_tr, y_tr, X_va, y_va, maximize=True, metric='auc'):
    fit_sig = inspect.signature(model.fit)
    if 'early_stopping_rounds' in fit_sig.parameters:
        model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False, early_stopping_rounds=300)
    else:
        try:
            from xgboost.callback import EarlyStopping
            cb = EarlyStopping(rounds=300, save_best=True, maximize=maximize, data_name='validation_0', metric_name=metric)
            model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False, callbacks=[cb])
        except Exception:
            model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)
    return model

def youden_threshold(y_true, y_prob):
    fpr, tpr, thr = roc_curve(y_true, y_prob)
    return float(thr[np.argmax(tpr - fpr)])

# --- version-proof RMSE helper ---
def rmse_score(y_true, y_pred):
    """Works across sklearn versions (with/without squared kwarg)."""
    try:
        return float(mean_squared_error(y_true, y_pred, squared=False))
    except TypeError:
        return float(np.sqrt(mean_squared_error(y_true, y_pred)))

# ---------------------------
# Leakage scan (quick single-feature probe)
# ---------------------------
def leakage_scan_and_drop(X, y_binary):
    bad_cols = []
    for col in X.columns:
        try:
            xx = X[[col]].copy()
            num = xx.select_dtypes(include=[np.number]).columns.tolist()
            cat = xx.select_dtypes(include=['object']).columns.tolist()
            steps = []
            if num:
                steps.append(('num', Pipeline([
                    ('imputer', SimpleImputer(strategy='median')),
                    ('scaler', StandardScaler())
                ]), num))
            if cat:
                steps.append(('cat', Pipeline([
                    ('imputer', SimpleImputer(strategy='most_frequent')),
                    ('onehot', _make_ohe(5))
                ]), cat))
            if not steps:
                continue
            pp = ColumnTransformer(steps, remainder='drop')
            xp = pp.fit_transform(xx)
            clf = LogisticRegression(max_iter=200)
            clf.fit(xp, y_binary)
            prob = clf.predict_proba(xp)[:,1]
            auc  = roc_auc_score(y_binary, prob)
            if auc >= 0.98:
                bad_cols.append(col)
        except Exception:
            continue
    if bad_cols:
        print(f" Potential leakage-like columns (AUC0.98 vs target): {bad_cols}")
        X = X.drop(columns=bad_cols, errors='ignore')
    return X, bad_cols

# ---------------------------
# BAGGED OOF frequency
# ---------------------------
def oof_frequency_bag(preprocessor, X, y, policy_tag):
    Xp = preprocessor.transform(X)
    spw, cw = compute_class_weights(y)

    all_oof, all_thr, seed_models = [], [], []

    for seed in BAG_SEEDS:
        skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=seed)
        oof_seed = np.zeros(len(y), dtype=float)
        models_seed, cals_seed = [], []

        for fold, (tr, va) in enumerate(skf.split(Xp, y), 1):
            X_tr, X_va = Xp[tr], Xp[va]
            y_tr, y_va = y[tr], y[va]

            lgbm = make_lgbm_classifier(cw)
            xgbc = make_xgb_classifier(spw)
            lgbm = fit_lgbm_es(lgbm, X_tr, y_tr, X_va, y_va)
            xgbc = fit_xgb_es(xgbc, X_tr, y_tr, X_va, y_va, maximize=True, metric='auc')

            p_blend = 0.5*lgbm.predict_proba(X_va)[:,1] + 0.5*xgbc.predict_proba(X_va)[:,1]
            ir = IsotonicRegression(out_of_bounds='clip')
            ir.fit(p_blend, y_va)
            p_cal = ir.predict(p_blend)

            auc = roc_auc_score(y_va, p_cal)
            print(f"[{policy_tag}] seed={seed} fold={fold} AUC={auc:.4f}")
            if auc >= 0.99:
                raise RuntimeError(f" ABORT: [{policy_tag}] seed {seed} fold {fold} AUC={auc:.4f}  leakage suspected")

            oof_seed[va] = p_cal
            models_seed.append((lgbm, xgbc))
            cals_seed.append(ir)

        auc_oof = roc_auc_score(y, oof_seed)
        thr = youden_threshold(y, oof_seed)
        print(f"[{policy_tag}] seed={seed} OOF AUC={auc_oof:.4f} | thr={thr:.4f}")

        all_oof.append(oof_seed)
        all_thr.append(thr)
        seed_models.append({'models': models_seed, 'calibrators': cals_seed})

    oof_mean = np.mean(all_oof, axis=0)
    auc_oof = roc_auc_score(y, oof_mean)
    thr_mean = float(np.mean(all_thr))
    print(f"[{policy_tag}] BAG OOF AUC={auc_oof:.4f} | avg thr={thr_mean:.4f}")

    return {'seeds': seed_models, 'threshold': thr_mean, 'auc_oof': float(auc_oof)}

def predict_frequency_bag(bundle, preprocessor, X):
    Xp = preprocessor.transform(X)
    probs_acc = np.zeros(Xp.shape[0], dtype=float)

    for seed_pack in bundle['seeds']:
        models, cals = seed_pack['models'], seed_pack['calibrators']
        seed_prob = np.zeros_like(probs_acc)
        for (lgbm, xgbc), ir in zip(models, cals):
            p = 0.5*lgbm.predict_proba(Xp)[:,1] + 0.5*xgbc.predict_proba(Xp)[:,1]
            seed_prob += ir.predict(p)
        seed_prob /= len(models)
        probs_acc += seed_prob

    probs_acc /= len(bundle['seeds'])
    return np.clip(probs_acc, 0.0, 1.0)

# ---------------------------
# BAGGED OOF severity (stack)
# ---------------------------
def oof_severity_bag(preprocessor, X, y_log, policy_tag):
    Xp = preprocessor.transform(X)
    all_rmse, seed_packs = [], []

    for seed in BAG_SEEDS:
        kf = KFold(n_splits=N_FOLDS, shuffle=True, random_state=seed)
        oof_stack = np.zeros((len(y_log), 3), dtype=float)
        fold_models = []

        for fold, (tr, va) in enumerate(kf.split(Xp), 1):
            X_tr, X_va = Xp[tr], Xp[va]
            y_tr, y_va = y_log[tr], y_log[va]

            m_t = make_lgbm_regressor_tweedie()
            m_x = make_xgb_regressor_phuber()
            m_l = make_lgbm_regressor_log()

            m_t = fit_lgbm_es(m_t, X_tr, y_tr, X_va, y_va)
            m_x = fit_xgb_es(m_x, X_tr, y_tr, X_va, y_va, maximize=False, metric='rmse')
            m_l = fit_lgbm_es(m_l, X_tr, y_tr, X_va, y_va)

            oof_stack[va, 0] = m_t.predict(X_va)
            oof_stack[va, 1] = m_x.predict(X_va)
            oof_stack[va, 2] = m_l.predict(X_va)

            fold_models.append((m_t, m_x, m_l))

        meta = Ridge(alpha=1.0, random_state=RANDOM_STATE)
        meta.fit(oof_stack, y_log)
        oof_pred = meta.predict(oof_stack)
        rmse = rmse_score(y_log, oof_pred)     # version-proof RMSE
        print(f"[{policy_tag}] seed={seed} OOF RMSE(log)={rmse:.4f}")

        all_rmse.append(rmse)
        seed_packs.append({'base_models': fold_models, 'meta': meta})

    rmse_mean = float(np.mean(all_rmse))
    print(f"[{policy_tag}] BAG OOF RMSE(log)={rmse_mean:.4f}")
    return {'seeds': seed_packs, 'rmse_oof_log': rmse_mean}

def predict_severity_bag(bundle, preprocessor, X):
    Xp = preprocessor.transform(X)
    preds_acc = np.zeros(Xp.shape[0], dtype=float)

    for seed_pack in bundle['seeds']:
        fold_models = seed_pack['base_models']
        meta = seed_pack['meta']
        base_stack = np.zeros((Xp.shape[0], 3), dtype=float)
        for (m_t, m_x, m_l) in fold_models:
            base_stack[:, 0] += m_t.predict(Xp)
            base_stack[:, 1] += m_x.predict(Xp)
            base_stack[:, 2] += m_l.predict(Xp)
        base_stack /= len(fold_models)
        preds_acc += meta.predict(base_stack)

    preds_acc /= len(bundle['seeds'])
    sev = np.expm1(preds_acc)
    return np.maximum(1.0, sev)

# ---------------------------
# MAIN
# ---------------------------
if __name__ == "__main__":
    print("="*60)
    print("=== PHASE 1: LEAK-PROOF DATA PREP (FREQUENCY-SEVERITY) ===")
    print("="*60)

    claims_history = load_csv(os.path.join(TRAIN_DIR, 'claims_history.csv'))
    feat_master    = create_feature_master(TRAIN_DIR)      # NO claims merge here
    targets        = build_targets_from_claims(feat_master, claims_history)
    X_train_full, y_pack = build_Xy(feat_master, targets)

    # simple preflight scan (optional)
    probe_target = y_pack['pet']
    X_train_full, dropped_cols = leakage_scan_and_drop(X_train_full, probe_target)
    if dropped_cols:
        print(f" Dropped suspicious columns: {dropped_cols}")

    preprocessor = get_preprocessor(X_train_full)
    preprocessor.fit(X_train_full)

    del claims_history, targets
    gc.collect()

    print("\n" + "="*60)
    print("=== PHASE 2: BAGGED OOF TRAINING (CPU) ===")
    print("="*60)

    bundles = {}
    metrics = {}

    for disp, key in [('Auto','auto'), ('Renters','renters'), ('Pet','pet')]:
        print(f"\n--- {disp} ---")
        y_freq = y_pack[key].astype(float)
        print(f"  -> P* positives: {int(y_freq.sum())}/{len(y_freq)}")

        freq_bundle = oof_frequency_bag(preprocessor, X_train_full, y_freq, f"P*_{disp}")
        bundles[f'p_{key}'] = freq_bundle
        metrics[f'auc_oof_p_{key}'] = freq_bundle['auc_oof']

        mask = (y_freq == 1.0)
        y_sev_log = y_pack[f'sev_log_{key}'][mask]
        print(f"  -> SEV* training size: {int(mask.sum())}")
        if mask.sum() > N_FOLDS:
            sev_bundle = oof_severity_bag(preprocessor, X_train_full.loc[mask], y_sev_log, f"SEV*_{disp}")
        else:
            class DummyReg:
                def predict(self, X): return np.zeros(X.shape[0], dtype=float)
            sev_bundle = {'seeds': [{'base_models': [], 'meta': DummyReg()}], 'rmse_oof_log': 0.0}
        bundles[f'sev_{key}'] = sev_bundle
        metrics[f'rmse_oof_log_sev_{key}'] = sev_bundle['rmse_oof_log']

    joblib.dump(bundles, 'bagged_oof_bundles.joblib')
    joblib.dump(preprocessor, 'bagged_oof_preprocessor.joblib')
    with open('bagged_oof_metrics.json', 'w') as f:
        json.dump(metrics, f, indent=2)
    print("\n Saved bagged OOF bundles, preprocessor, and OOF metrics.")
    print("OOF metrics:", metrics)

    print("\n" + "="*60)
    print("=== PHASE 3: PREDICT & SUBMIT ===")
    print("="*60)

    feat_test = create_feature_master(TEST_DIR)
    if 'user_id' not in feat_test.columns:
        raise ValueError("user_id missing in test data")
    X_test = feat_test.drop(columns=['user_id'], errors='ignore')

    expected_cols = X_train_full.columns.tolist()
    X_test_aligned = X_test.reindex(columns=expected_cols, fill_value=np.nan)

    submission = pd.DataFrame({'user_id': feat_test['user_id']})
    for key in ['auto','renters','pet']:
        prob = predict_frequency_bag(bundles[f'p_{key}'], preprocessor, X_test_aligned)
        submission[f'p_{key}'] = np.clip(prob, 0.001, 0.999)
        sev  = predict_severity_bag(bundles[f'sev_{key}'], preprocessor, X_test_aligned)
        submission[f'sev_{key}'] = sev

    cols = ['user_id','p_auto','p_renters','p_pet','sev_auto','sev_renters','sev_pet']
    submission[cols].to_csv('submission_predictions_leakproof.csv', index=False)

    print("\n Created 'submission_predictions_leakproof.csv'. Head:")
    print(submission[cols].head())
    print("\n Done. Leak-proof, calibrated, regularized, and OOF-evaluated.")
```
